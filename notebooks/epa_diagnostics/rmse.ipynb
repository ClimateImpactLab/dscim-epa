```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
from statsmodels.tools.eval_measures import rmse
```

```python
year=2100
ssp='SSP2'
model='OECD Env-Growth'
```

```python
points = pd.read_csv(
    "/mnt/CIL_integration/menu_results_AR6_fixed_power/AMEL_clipped_ax2/2020/unmasked/risk_aversion_constant_eta2.0_rho0.0_damage_function_points.csv"
)
subset = points.loc[(points.model==model) & (points.ssp==ssp) & (points.year >= year-2) & (points.year <= year+2)]
```

# y=ax + bx^2

```python
#define figure size
fig = plt.figure(figsize=(12,8))

#produce regression plots
model = ols('damages ~ -1 + anomaly + np.power(anomaly, 2)', data=subset).fit()
fig = sm.graphics.plot_regress_exog(model, 'np.power(anomaly, 2)', fig=fig)
print(model.summary())
# now generate predictions
ypred = model.predict(exog={'anomaly' : subset.anomaly})

# calc rmse
RMSE = rmse(subset.damages[[i for i in subset.damages.index if i in ypred.index]], 
            ypred)
print('RMSE', RMSE)

plt.savefig("/mnt/CIL_integration/bubble_diagnostics/fit_plots/ax_bx2.png", dpi=300, bbox_inches='tight')
```

# y=ax^2

```python
#define figure size
fig = plt.figure(figsize=(12,8))

#produce regression plots
model = ols('damages ~ -1 + np.power(anomaly, 2)', data=subset).fit()
fig = sm.graphics.plot_regress_exog(model, 'np.power(anomaly, 2)', fig=fig)
print(model.summary())
# now generate predictions
ypred = model.predict(exog=dict(anomaly=subset.anomaly))

# calc rmse
RMSE = rmse(subset.damages[[i for i in subset.damages.index if i in ypred.index]], 
            ypred)
print('RMSE', RMSE)

plt.savefig("/mnt/CIL_integration/bubble_diagnostics/fit_plots/bx2.png", dpi=300, bbox_inches='tight')
```
