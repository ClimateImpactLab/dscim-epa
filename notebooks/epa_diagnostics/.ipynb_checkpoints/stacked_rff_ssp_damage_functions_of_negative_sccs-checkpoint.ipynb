```python
import xarray as xr
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from dscim.diagnostics.damage_function import damage_function
```

```python
# sectors=['agriculture_clipped', 'mortality_clipped', 'energy_clipped', 'labor_clipped', 'AMEL_clipped', "CAMEL_clipped"]
sectors=['CAMEL_clipped']

eta_rhos = {
    2.0 :0.0,
    # 1.016010255: 9.149608e-05,
    # 1.244459066: 0.00197263997,
    # 1.421158116: 0.00461878399,
    # 1.567899395: 0.00770271076,
}

v=3
rff=False
```

```python
def plot_stacked(sector,
                 recipe='risk_aversion',
                disc='constant',
                eta=2.0,
                 rho=0.0,
                 xlim=(-1, 8),
                years = [2020, 2050, 2090, 
                         2100, 2200, 2300],
                 sharey=False,
                ):

    root_rff=f'/mnt/CIL_integration/rff3_with_disc_factors/iter0-19/{sector}/2020/unmasked_None'
    root_ssp=f'/mnt/CIL_integration/menu_results_AR6_bc39_epa_vsl_continuous/{sector}/2020/unmasked'

    # finding tiny SCCs for RFF
    sccs = xr.open_dataset(
        f"{root_rff}/{recipe}_{disc}_eta{eta}_rho{rho}_uncollapsed_sccs.nc4"
    ).sel(weitzman_parameter='0.5').uncollapsed_sccs
    cutoff=sccs.chunk({'simulation': -1}).quantile(0.0001, ['rff_sp', 'simulation'])
    sccs = sccs.where(sccs < cutoff, drop=True)
    rff_ids = sccs.rff_sp.values

    # overlap ssp and rff
    ds_rff = xr.open_dataset(f"{root_rff}/{recipe}_{disc}_eta{eta}_rho{rho}_damage_function_coefficients.nc4").sel(rff_sp=rff_ids, year=years)
    ds_ssp = xr.open_dataset(f"{root_ssp}/{recipe}_{disc}_eta{eta}_rho{rho}_damage_function_coefficients.nc4").sel(year=years)

    temps = xr.DataArray(
        np.arange(xlim[0], xlim[1], 0.1),
        coords={
            'anomaly' : np.arange(xlim[0], xlim[1], 0.1)
        })

    fit_rff = ds_rff['anomaly'] * temps + ds_rff['np.power(anomaly, 2)'] * temps ** 2
    fit_rff = fit_rff.to_dataframe('fit').reset_index()
    fit_rff['model'] =  "rff_sp: " + fit_rff['rff_sp'].astype(str)
    fit_rff = fit_rff[['fit','year','model','anomaly']]

    fit_ssp = ds_ssp['anomaly'] * temps + ds_ssp['np.power(anomaly, 2)'] * temps ** 2
    fit_ssp = fit_ssp.to_dataframe('fit').reset_index()
    fit_ssp["model"] = fit_ssp.ssp + "-" + fit_ssp.model
    fit_ssp = fit_ssp[['fit','year','model','anomaly']]

    if rff==True:
        fit = pd.concat([fit_ssp, fit_rff]).set_index(['year','model','anomaly'])
        pal=sns.color_palette("Paired", 6) + sns.color_palette("Greys", len(rff_ids))
    else:
        fit = fit_ssp.set_index(['year','model','anomaly'])
        pal=sns.color_palette("Paired", 6)

    g = sns.relplot(
        data=fit,
             x='anomaly',
             y='fit',
             hue='model',
             col='year',
             col_wrap=3,
             kind='line',
             palette = pal,
             facet_kws={'sharey': sharey, 'sharex': True},
             legend='full',
            )

    g.fig.suptitle(f"{sector} {recipe} {disc}")

    plt.savefig(f'/mnt/CIL_integration/rff_diagnostics/v{v}/damage_functions_negative_SCCs/rff_ssp_stacked/stacked_{sector}_{recipe}_{disc}_{eta}_{rho}_xlim{xlim}_rff-{rff}_years{years}.png', 
                bbox_inches='tight', 
                dpi=300
               )
```
